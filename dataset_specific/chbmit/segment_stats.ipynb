```python
import os
import json
import pandas as pd

fs = 256 # Sample Rate
segment_dir = os.path.join('..', 'segment_clean')
```

```python
with os.scandir(segment_dir) as it:
    for entry in it:
        if not entry.name.startswith('.') and entry.is_dir():
            subtitle = entry.name
            seg_cnt = []; seg_len = []
            with os.scandir(entry.path) as in_it: # Traverse each subfolder
                for pat_dir in in_it:
                    if not pat_dir.name.startswith('.') and pat_dir.is_dir():
                        patstr = pat_dir.name
                        with open(os.path.join(pat_dir.path, 'segment_info.json'), 'r') as f:
                            seg_lst = json.load(f)
                        
                        pat_seg_cnt = {'Subject': patstr, 'Inter': 0, 'Pre': 0, 'Onset': 0}
                        pat_seg_len = {'Subject': patstr, 'Inter': 0, 'Pre': 0, 'Onset': 0}
                        for seg in seg_lst:
                            for label in ('inter', 'pre', 'onset'):
                                if label in seg["Label"].lower():
                                    pat_seg_cnt[label.capitalize()] += 1
                                    pat_seg_len[label.capitalize()] += (seg["Span"][1] - seg["Span"][0]) / (3600 * fs) # Display values in hours
                                    break
                        
                        seg_cnt.append(pat_seg_cnt); seg_len.append(pat_seg_len)
            df_cnt = pd.DataFrame(seg_cnt).set_index('Subject').sort_index(); df_len = pd.DataFrame(seg_len).set_index('Subject').sort_index()
            df_len['Ratio'] = df_len['Inter'] / df_len['Pre']
            print('## '+subtitle, df_cnt.to_markdown(), '\n', df_len.to_markdown(), sep='\n')
            
```
